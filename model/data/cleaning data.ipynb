```python
#Este For se realiza para limpiar la Data obtenida de Tweepy
#Debido que al traer la Data se obtiene información que no esta relacionada a los nombres de los municipios.
#Se crea una bandera para comenzar a limpiar unicamente los que son municipios.
lista = []
flag = False
for i in range(df.shape[0]):
    value = df['location'][i].replace('.', '')
    for j in range(len(Array_Municipios)):
        if value == Array_Municipios[j]:
            lista.append(True)
            flag = True
    if flag == False:
        lista.append(False)
    flag = False
```

```python
#Se crea la columna flag con la lista creada anteriormente
df['Flag'] = lista
```

```python
#Se deja unicamente los registros en los cuales la bandera es igual a TRUE.
df_clean = df.query("Flag == True")
```

```python
#Se crea una nueva lista con los Municipios obtenidos.
Final_list = []
for i in range(df_clean.shape[0]):
    Final_list.append(df_clean['location'][i])
```

```python
#Se transforma dicha lista en un DataFrame para realizar su Post procesamiento.
df_Final_pre = pd.DataFrame(Final_list)
```

```python
#Se limpian los objetos de la lista.
Final_list = []
for i in range(df_Final_pre.shape[0]):
    value = df_Final_pre[0][i].replace('.', '')
    Final_list.append(value)
```

```python
df_Final = pd.DataFrame(Final_list)
new_df =  pd.DataFrame(Array_Municipios)
```

```python
#Se rellena de 0 los contadores
zero = 0
new_df['Count'] = zero
```

```python
#En este doble bucle for se realiza la comparación y si se encuentra dicho objeto de la lista.
#Se va sumando un +1 en la posicion que se encuentra.
for i in range(df_Final.shape[0]):
    for j in range(new_df.shape[0]):
        value_1= new_df[0][j]
        value_2 = df_Final[0][i]
        if value_1 == value_2:
            count = new_df['Count'][j]
            count +=1
            new_df.at[j, 'Count'] = count 
```
