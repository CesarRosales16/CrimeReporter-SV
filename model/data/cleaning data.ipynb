```python
df['Flag'] = lista
```

```python
df_clean = df.query("Flag == True")
```

```python
Final_list = []
for i in range(df_clean.shape[0]):
    Final_list.append(df_clean['location'][i])
```

```python
df_Final_pre = pd.DataFrame(Final_list)
```
